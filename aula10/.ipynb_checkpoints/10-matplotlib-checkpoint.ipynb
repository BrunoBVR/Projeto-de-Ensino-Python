Matplotlib é provavelmente o pacote Python mais utilizado para gráficos 2D. Ele fornece uma maneira muito rápida de visualizar os dados de Python e figuras de qualidade de publicação em vários formatos. Vamos explorar matplotlib no modo interativo que cobre casos mais comuns.
    
Esta é uma das aulas mais interessantes até agora por um motivo muito simples: visualização!
Visualizar o que estamos fazendo dá corpo à tudo que vimos até agora, além de ser bem bonito ;)

A documentação completa do módulo esta na página [matplotlib.org](https://matplotlib.org/).

A primeira coisa a ser feita é importar o módulo para ter acesso a seu conteúdo:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Vamos utilizar bastante o módulo `numpy`, já que ele é chave para lidar com dados numéricos (que é o que queremos para gerar gráficos).

A função mais básica do `matplotlib` é a plotagem de sequências numéricas. Em python, essas sequências podem ser listas ou `arrays` do `numpy`.

Por exemplo, vamos criar um gráfico (plotar) uma sequência de números.

```python
# Criar a sequência:
seq = [1,2,3,4,5,6]

# Plotando com plt
plt.plot(seq)

# Mostrando o gráfico
plt.show()
```

Por *default*, se apenas uma sequência é passada, o matplotlib a interpreta como variável dependente (eixo y) e assume que o eixo x é uma sequência ordenada.

```python
# Mais um exemplo
seq2 = [1,4,9,16,25,36]

plt.plot(seq2)
plt.show()
```

Se passarmos explicitamente duas sequências, ele interpreta a primeira como contendo os valores do eixo x e a segunda, do eixo y:

```python
seq_x = [2,4,6,8,10,12]
seq_y = [4,16,36,64,100,144]

plt.plot(seq_x, seq_y)
plt.show()
```

Podemos usar algumas ferramentas para personalizar o nosso gráfico

Podemos plotar funções do `numpy`, por exemplo:

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.show()
```

Lembrando que a função `linspace` gera uma sequência de números igualmente espaçados (no caso, geramos 100 números entre 0 1 10).

Podemos personalizar um pouco nossas curvas com algumas opções do módulo.

Após informar os valores a serem plotados, podemos inserir uma sequênica, sempre entre '  ', de até três informações na ordem:
`[marcador][linha][cor]`.

Vamos ver alguns exemplos:

```python
# Usando pontos em vez de linha

plt.plot(x, np.sin(x), '.')
plt.show()
```

```python
# Usando círculos em vez de linha

plt.plot(x, np.sin(x),'o')
plt.show()
```

```python
# Usando quadrados em vez de linha ('s'quare)

plt.plot(x, np.sin(x), 's')
plt.show()
```

```python
# Usando círculos vermelhos (o - círculo, r - vermelho)

plt.plot(x, np.sin(x), 'or')
plt.show()
```

```python
# Usando linha, pontos e a cor vermelha

plt.plot(x, np.sin(x), '.-r')
plt.show()
```

Para ver todas as opções acessem [matplotlib.org - info](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html)

Podemos plotar várias curvas no mesmo gráfico. Basta usar o comando `plot` para cada curva desejada e finalizar com o `plt.show()`:


```python
x = np.linspace(0, 2, 21) # 21 pontos no intervalo [0, 2]
plt.plot(x, x)            # função y = x
plt.plot(x, x**2)         # função y = x^2
plt.plot(x, x**3)         # função y = x^3
plt.show()
```

O módulo automaticamente ajusta as escalas e as cores diferentes para diferentes chamadas do `plot()`. Podemos personalizar bastante:

```python
x = np.linspace(0, 2, 21)
plt.plot(x, x, 'b--')   # linha azul (b) tracejada (--)
plt.plot(x, x**2, 'go') # nenhuma linha, bolinhas (o) verdes (g)
plt.plot(x, x**3, 'r^') # nenhuma linha, triângulos (^) vermelhos (r)
plt.show()
```

Vamos deixar o gráfico o mais comunicativo possível (afinal, imagens servem para resumir de maneira rápida e eficiente um resultado!).

Vamos usar algumas opções novas:
* Colocar legendas para difenreciar as curvas (usamos o parâmetro `label` dentro da função `plot` e a função `legend` ao final de todos os `plot()`.
* Título do gráfico (usaremos a função `title` após todos os `plot()`
* Colocaremos descrições nos eixos (funções `xlabel()` e `ylabel()`
* Usaremos a função `grid` para desenhar uma malha quadriculada na figura.

Acompanhe os detalhes no exemplo abaixo:

```python
x = np.linspace(0, 2, 21)
plt.plot(x, x, 'b--', label='linear')
plt.plot(x, x**2, 'go', label='quadrático')
plt.plot(x, x**3, 'r^', label='cúbico')
plt.xlabel('teste x')
plt.ylabel('teste y')
plt.title('Gráfico simples')
plt.legend()
plt.grid()
plt.show()
```

Olhem como fica fofo nosso gráfico! COm poucos comandos podemos deixar nossos resultados apresentáveis e claros (nunca esqueçam da importância da apresentação dos resultados de forma clara. Eu sei e concordo que fazer contas é maravilhoso e as equações são as coisas mais belas do universo, mas nem todos têm esse esclarecimento...)

### Plotagem paramétrica

Rapidamente vou mostrar que podemos plotar curvas parametrizadas com a função `plot()`. Por exemplo, uma elipse centrada na origem pode ser obtida da seguinte forma:

```python
t = np.linspace(0, 2*np.pi, 100)
x, y = 2*np.cos(t), np.sin(t)
plt.axis('equal')   # iguala as escalas horizontal e vertical
plt.plot(x, y)
plt.show()
```

## Escolhendo os limites dos eixos

Vejam as seguintes curvas:

```python
x = np.linspace(0, 2, 61)
plt.plot(x, x)
plt.plot(x, 1 / x)
plt.plot(x, 1 / x**2)
plt.show()
```

Obtemos um pequeno aviso (warning) ao fazer essa plotagem pois as funções que usamos possuem possíveis divisões por zero (1/x e 1/x²). Além disso, nesse exemplo em específico, as três curvas possuem um ponto de interseção interessante em x = 1 que não está bem visível no=a figura.

Podemos escolher os limites dos eixos com as funções `xlim()` e `ylim():

```python
x = np.linspace(0, 2, 61)
plt.plot(x, x)
plt.plot(x, 1 / x)
plt.plot(x, 1 / x**2)
plt.xlim(0.1, 2)  # de 0.1 até 2 na horizontal
plt.ylim(0, 4)  # de 0 até 4 na vertical
plt.show()
```

Bem melhor!

Existe uma opção de plotagem chamada `scatter`, usada para fazer gráficos de dispersão. Essencialmente, é uma ferramenta que plota pontos em vez de curvas. Veja um exemplo:

```python
x = np.linspace(0,10,100)
plt.scatter(x, np.cos(x))
plt.show()
```

## Criando subplotagens (subplots)

Podemos criar uma única figura com muitos gráficos (diferente de um único gráfico com muitas curvas!). Para tal, usamos a opção `subplots`. O uso pode parecer confuso de início, mas os resultados são ótimos. Acompanhem com muito cuidado o passo a passo e tentem entender a razão de cada pedaçõ do código.

```python
x = np.linspace(0,10,10)

# Criamos dois objetos: a figura (fig) e os pares de eixos (ax)
# Aqui vamos criar subplots com 2 linhas e duas colunas.
# Pensem como se a figura fosse uma grande matriz e cada
# entrada da matriz é um gráfico.
# Assim, teremos uma figura na posição (0,0), uma na (0,1)
# uma na (1,0) e a última na (1,1)

# Finalmente, usamos figsize para definir o tamanho da figura
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))
# Agora podemos plotar em cada um dos gráficos.
# Em vez de usarmos o plt, usamos a posição da figura
# que pode ser de [0,0] a [1,1]
ax[0, 0].plot(x, x/2, '^')
ax[0,1].scatter(x, np.cos(x))
ax[1,0].plot(x, np.exp(x), 'r-o')
ax[1,1].plot(x, np.exp(-x)*np.sin(x), 'g-o')
```
